```python
#get the correct package for data structures
#link/load the raw data files


import pandas as pd

#Peru
TPdata="https://github.com/vzakhozhyi/599-A-Final-Project/raw/master/Data%20Original/Temp_Peru.xls"
TP=pd.read_excel(TPdata)
```

```python
TP.shape
```

```python
TP
#check it out! 
```

```python
#drop columns we don't need-- the columns with irrelevant and/or NaN values
Tclean=TP.drop(columns=[' ISO3', ' ISO2'])
```

```python
#check the shape to see if those columns got dropped. 
Tclean.shape
```

```python
#re-name the column headers
#we do this by first creating a list with the names we want
#then use a zip function to zip these values into a dict where the name we want references the names we want to replace. 
#then replace, via a rename function, the values we want in place of the values we want to replace
#remember to do this inplace! 

Tnewnames=['temperature', 'year', 'month', 'country']
Tnew={old:new for old,new in zip(Tclean,Tnewnames)}
Tclean.rename(columns=Tnew,inplace=True)
Tclean
```

```python
#replace the PER values with Peru
Tclean['country']=Tclean['country'].replace(['PER'], 'Peru')
```

```python
#limit the data to only inlcude the years of interest, 2006-2015
years=['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']
Tclean = Tclean[Tclean.year.isin(years)]
Tclean
```

```python
#reset the index to start at 0!
Tclean.reset_index(drop=True)
```

```python
#Create a new file out of our cleaned data and we're done
Tclean.to_csv("PeruTemperatureFinal.csv",index=None)
```
