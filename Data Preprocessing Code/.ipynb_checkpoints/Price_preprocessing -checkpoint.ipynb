## Price Dataset


```python
import pandas as pd

CsvFile='https://raw.githubusercontent.com/vzakhozhyi/599-A-Final-Project/master/Data%20Original/wfp_market_food_prices.csv'

dataCsv=pd.read_csv(CsvFile, header= 0, encoding = "ISO-8859-1")
```

View each column and data type

```python
dataCsv.info()
```

Find each country in dataset and search for any input errors

```python
dataCsv.adm0_name.unique()
```

Let's see what the dataset looks like.

```python
dataCsv.head(10)
```

```python
dataCsv.tail(10)
```

Subset the dataset. Define our target country (Peru) and crop (Potatoes).

```python
TargetCountry = [ "Peru"] 
```

```python
PeruPrice = dataCsv[dataCsv.adm0_name.isin(TargetCountry)]
```

Show unique values of crops in each target countries.

```python
PeruPrice.cm_name.unique()
```

```python
TargetCrop = ["Potatoes"]
```

```python
PeruPrice = PeruPrice[PeruPrice.cm_name.isin(TargetCrop)]
```

Delete columns that will not be used in our analysis.

```python
# This is what we want get rid of:
ByeColumns = ['adm0_id', 'adm1_id','mkt_id','cm_id','cur_id','pt_id','um_id']#,'mp_commoditysource']

#this would change the original: GhanaYamPrice.drop(byeColumns,axis=1,inplace=False)
PeruPrice=PeruPrice.drop(ByeColumns,axis=1, inplace=False).reset_index(drop=True)
```

```python
PeruPrice.tail(10)
```

Our data has the potato prices by market in a given month. We instead want the average price of Potatoes by country (Peru) in a given year and month.

```python
PriceData = PeruPrice.groupby(['adm0_name','cm_name','mp_year','mp_month'], as_index=False)[['mp_price']].mean()
```

data.groupby('month', as_index=False).agg({"duration": "sum"}), The groupby output will have an index or multi-index on rows corresponding to your chosen grouping variables. To avoid setting this index, pass “as_index=False” to the groupby operation. This function is essentially taking all the market level price data for the corresponding year AND month and calculating the average.



```python
list(PriceData)
```

PriceData.columns will replace the column names in the order listed.

```python
PriceData.columns=['country','crop','year','month','price']
```

```python
PriceData.tail(10)
```

Next we want to include price data from years 2006 to 2015.

```python
TargetYear = range(2006,2016)
PriceData = PriceData[PriceData.year.isin(TargetYear)]
```

```python
PriceData.head(10)
```

```python
PriceData.tail(10)
```

Saving files to R:

```python
PriceData.to_csv("PeruPriceFinal.csv",index=None)
```

# Merge the datasets

Load cleaned datasets:

```python
pricelink = "https://raw.githubusercontent.com/vzakhozhyi/599-A-Final-Project/master/Data%20Final/PeruPriceFinal.csv"
templink = "https://raw.githubusercontent.com/vzakhozhyi/599-A-Final-Project/master/Data%20Final/PeruTemperatureFinal.csv"
rainfalllink = "https://raw.githubusercontent.com/vzakhozhyi/599-A-Final-Project/master/Data%20Final/PeruRainfallFinal.csv"


PriceFinal = pd.read_csv(pricelink,header=0)
TempFinal = pd.read_csv(templink,header=0)
RainfallFinal = pd.read_csv(rainfalllink,header=0)
```

Check dataset before merge:

```python
list(PriceFinal)
```

```python
list(TempFinal)
```

```python
list(RainfallFinal)
```

Merge three dataset:

```python
DataFinal = pd.merge(PriceFinal, TempFinal, how='left', on=["country","year","month"])
```

```python
DataFinal = pd.merge(DataFinal, RainfallFinal, how='left', on=["country","year","month"])
```

Export final dataset:

```python
DataFinal.to_csv("DataFinal.csv",index=None)
```
